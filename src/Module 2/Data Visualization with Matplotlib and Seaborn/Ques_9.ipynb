```python
# Ques_9.ipynb

import matplotlib.pyplot as plt

# Data for bar plot
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]

# Data for scatter plot
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Subplot 1: Bar plot
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Product Sales')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales (in thousands)')
axs[0].grid(True, linestyle='--', alpha=0.5)

# Subplot 2: Scatter plot
axs[1].scatter(heights, weights, color='orange', s=100)
axs[1].set_title('Height vs Weight')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')
axs[1].grid(True, linestyle='--', alpha=0.5)

# Adjust layout and show
plt.tight_layout()
plt.show()
```
