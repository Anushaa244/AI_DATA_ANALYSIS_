```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.random.randint(1, 10, size=5)

# Create a figure and an array of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plotting on the first subplot (top-left)
axs[0, 0].plot(x, y, label='Sine Wave', color='b')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].legend()

# Plotting on the second subplot (top-right)
axs[0, 1].scatter(x, y2, color='r', label='Cosine Scatter')
axs[0, 1].set_title('Cosine Scatter')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].legend()

# Plotting on the third subplot (bottom-left)
axs[1, 0].bar(np.arange(5), y3, color='g', label='Random Bar')
axs[1, 0].set_title('Random Bar Plot')
axs[1, 0].set_xlabel('Index')
axs[1, 0].set_ylabel('Value')
axs[1, 0].legend()

# Hide the fourth subplot (bottom-right) as an empty plot
axs[1, 1].axis('off')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the combined figure with subplots
plt.show()
```
