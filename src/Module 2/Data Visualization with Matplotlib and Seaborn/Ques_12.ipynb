```python
# Step 1: Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load a sample dataset
# The 'tips' dataset is available in Seaborn and contains information about restaurant tips.
tips = sns.load_dataset('tips')

# Step 3: Create a pairplot
# A pairplot will show pairwise relationships in the dataset, and we use the 'sex' column to color the data points.
sns.pairplot(tips, hue='sex', palette='Set1')

# Step 4: Show the plot
plt.show()
```
