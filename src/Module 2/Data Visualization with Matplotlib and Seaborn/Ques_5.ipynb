```python
# Ques_5.ipynb

# Step 1: Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Define the data
data = {
    'Height (cm)': [160, 170, 165, 180, 175],
    'Weight (kg)': [55, 65, 60, 80, 76]
}

df = pd.DataFrame(data)

# Step 3: Create the scatter plot using Seaborn
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Height (cm)', y='Weight (kg)', data=df, color='purple', s=100)

# Step 4: Add titles and labels
plt.title('Height vs Weight (Seaborn Scatter Plot)', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Optional: Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Step 5: Show the plot
plt.show()
```
