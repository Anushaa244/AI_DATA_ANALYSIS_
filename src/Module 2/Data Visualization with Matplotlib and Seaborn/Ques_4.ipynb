```python
# Step 1: Import seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Define the data
data = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Sales': [23, 75, 57, 90, 65]
}

df = pd.DataFrame(data)

# Step 3: Create the Seaborn bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Product', y='Sales', data=df, palette='Blues')

# Step 4: Add titles and labels
plt.title('Sales by Product (Seaborn Bar Plot)', fontsize=16)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales (in thousands)', fontsize=12)

# Step 5: Show the plot
plt.show()
```
