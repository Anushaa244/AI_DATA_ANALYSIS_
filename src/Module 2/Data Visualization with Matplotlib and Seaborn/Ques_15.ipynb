```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the iris dataset
data = sns.load_dataset('iris')

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=[float, int])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Set the size of the figure
plt.figure(figsize=(10, 8))

# Create a heatmap using Seaborn
sns.heatmap(corr_matrix,
            annot=True,  # Display the correlation values in each cell
            cmap='coolwarm',  # Color palette
            center=0,  # Center the color scale at 0
            linewidths=1,  # Linewidth between cells
            linecolor='gray',  # Line color
            cbar_kws={"shrink": 0.8},  # Shrink the color bar
            fmt='.2f',  # Format the correlation values to 2 decimal places
            square=True)  # Make the heatmap square

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16)

# Show the plot
plt.show()
```
