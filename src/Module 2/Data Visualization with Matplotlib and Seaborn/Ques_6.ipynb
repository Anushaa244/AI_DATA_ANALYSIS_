```python


# Step 1: Import matplotlib
import matplotlib.pyplot as plt

# Step 2: Define data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]  # in thousands

# Step 3: Customize plot style and create the bar plot
plt.figure(figsize=(8, 5))
plt.bar(products, sales, color='orange', edgecolor='black')

# Step 4: Add title and axis labels with custom font size
plt.title('Customized Sales Bar Plot', fontsize=16, fontweight='bold')
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales (in thousands)', fontsize=12)

# Step 5: Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optional: Add value labels on top of each bar
for i, value in enumerate(sales):
    plt.text(i, value + 2, str(value), ha='center', fontsize=10, fontweight='bold')

# Step 6: Show the plot
plt.tight_layout()
plt.show()
```
