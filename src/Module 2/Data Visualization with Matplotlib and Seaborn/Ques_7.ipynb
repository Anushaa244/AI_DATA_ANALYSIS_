```python
# Ques_7.ipynb

# Step 1: Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 2: Define the data
data = {
    'Height (cm)': [160, 170, 165, 180, 175],
    'Weight (kg)': [55, 65, 60, 80, 76],
    'Size': [10, 15, 13, 18, 12]
}

df = pd.DataFrame(data)

# Step 3: Create the advanced scatter plot with size variation
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='Height (cm)',
    y='Weight (kg)',
    size='Size',
    sizes=(100, 400),  # scale marker sizes
    data=df,
    color='teal',
    legend=False
)

# Step 4: Add titles and labels
plt.title('Height vs Weight with Point Size Based on Data', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Step 5: Show the plot
plt.tight_layout()
plt.show()
```
