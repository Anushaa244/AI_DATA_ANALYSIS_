```python
# Step 1: Import required libraries
import pandas as pd
import seaborn as sns

# Step 2: Load the dataset
# Here we load the 'iris' dataset from seaborn. Replace this with your dataset if needed.
iris = sns.load_dataset('iris')

# Step 3: Group the data by species and compute the mean of each feature
grouped_data = iris.groupby('species').mean()

# Step 4: Display the result
print(grouped_data)
```
