```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns

# Load the Iris dataset using seaborn
data = sns.load_dataset('iris')

# Convert the relevant columns (numerical ones) into a NumPy array
numeric_data = data.select_dtypes(include=[float, int]).values

# Calculate the mean of each column using NumPy
mean_values = np.mean(numeric_data, axis=0)

# Print the mean values for each numerical feature
print("Mean of each feature in the dataset:")
for feature, mean in zip(data.select_dtypes(include=[float, int]).columns, mean_values):
    print(f"{feature}: {mean:.2f}")

# Additional basic operations using NumPy
# Calculate the standard deviation
std_devs = np.std(numeric_data, axis=0)
print("\nStandard Deviation of each feature:")
for feature, std in zip(data.select_dtypes(include=[float, int]).columns, std_devs):
    print(f"{feature}: {std:.2f}")

# Calculate the median of each column
median_values = np.median(numeric_data, axis=0)
print("\nMedian of each feature:")
for feature, median in zip(data.select_dtypes(include=[float, int]).columns, median_values):
    print(f"{feature}: {median:.2f}")

# Calculate the sum of each column
sum_values = np.sum(numeric_data, axis=0)
print("\nSum of each feature:")
for feature, sum_val in zip(data.select_dtypes(include=[float, int]).columns, sum_values):
    print(f"{feature}: {sum_val:.2f}")
```
